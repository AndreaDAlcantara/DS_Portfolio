# **Exploratory Data Analysis - CVS files**

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/SCPRC-EST2022-18+POP.csv")
df
```

Drop rows with NaN values.

```python
df = df.dropna(how='all')
df = df.dropna(how='all',axis='columns')
df = df.dropna(thresh=len(df.columns)-1)
df
```

Turn 1st row into column names.

```python
header_row = df.iloc[0]
df2 = pd.DataFrame(df.values[1:], columns=header_row)
df2
```

Rename columns.

```python
df2.columns = ['Region','Total_Pop','Total_Pop_Over18','Perc_Over18']
df2
```

Remove character "." from the Region column.

```python
df2['Region'] = df2['Region'].str.replace('.','')
df2
```

Turn Population columns into float type in order to be used in the graphs.

```python
df2['Total_Pop'] = df2['Total_Pop'].str.replace(',', '').astype(float)
df2['Total_Pop_Over18'] = df2['Total_Pop_Over18'].str.replace(',', '').astype(float)
df2
```

Create dataset to work only with the States data.

```python
df3 = df2.iloc[5:57]
df3
```

```python
from jupyter_dash import JupyterDash
import dash
import plotly.graph_objects as go
import plotly.express as px
from dash import Dash, dcc, html, Input, Output, dash_table
```

```python
! pip install dash-bootstrap-components
import dash_bootstrap_components as dbc
```

```python
JupyterDash.infer_jupyter_proxy_config()
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
filtered_df = df3
#app = JupyterDash(__name__, external_stylesheets=external_stylesheets)
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.LUX])
#app = dash.Dash(external_stylesheets=[dbc.themes.LUX])

app.layout = html.Div([
    dbc.Row(dbc.Col(html.Div(html.H4('US Population by State')))),
    dbc.Row(dbc.Col(html.Div(
      dcc.Dropdown(id='dropdown',
          options=[{'label': 'Total Population', 'value': 'Total_Pop'},{'label': 'Population Over 18', 'value': 'Total_Pop_Over18'}],
          placeholder="Select",
          value='Total_Pop', clearable=False
      ))        
    ,width=6)),

    html.Br(),
    dbc.Row([dbc.Col(html.Div(dcc.Graph(id="graph")),width={"size": 5, "offset": 0}),dbc.Col(html.Div(dcc.Graph(id="graph2")),width={"size": 7, "offset": 0})],align="start")
])

@app.callback(
    Output('graph', 'figure'), 
    Output('graph2', 'figure'), 
    Input('dropdown', 'value')
)

def generate_chart(dropdown_value):
    pop_df = filtered_df
    if dropdown_value == 'Total_Pop':
      values = pop_df['Total_Pop']
      names = pop_df['Region']
      title_ = 'Total Population by State'
      fig = px.pie(values=values, names=names)

    if dropdown_value == 'Total_Pop_Over18':
      values = pop_df['Total_Pop_Over18']
      names = pop_df['Region']
      title_ = 'Population Over 18 by State'
      fig = px.pie(values=values, names=names)

    fig.update_layout(
        autosize=True,
        height=600
    )
    #fig.update_layout(margin = dict(t=30, l=20, r=20, b=30), )
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(showlegend=False)
    fig.update_layout(title=title_)

    values1 = pop_df['Total_Pop']
    values2 = pop_df['Total_Pop_Over18']
    names1 = pop_df['Region']
    fig2 = px.bar(pop_df, x=names1, y=[values1,values2], orientation='v', barmode='group')
    fig2.update_layout(
        title='US Total Population x Population Over 18',
        legend={'title_text':''},
        width=800,
        height=600,
        yaxis=dict(title='Population')
    )
    fig2.update_layout(legend=dict(
          yanchor="top",
          y=0.99,
          xanchor="right",
          x=0.99
    ))

    return fig, fig2   
```

```python
app.run_server(mode="inline")
```
